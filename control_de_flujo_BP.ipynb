<b><font color='MediumSeaGreen'>5.</font></b> Utilizando la misma base <span style='background:gainsboro'><font color='dimgray'>**flights.csv**</font></span> debes hacer un loop y clasificar los meses con una cantidad de pasajeros menor a la media. Para ello:




- Genera un nuevo objeto que represente la media de <span style='background:gainsboro'><font color='dimgray'>passengers</font></span>.

- Genera una columna en la base de datos que se llame <span style='background:gainsboro'><font color='dimgray'>underperforming</font></span> y asígnele 0.

- Ejecuta un loop que recorra cada observación de <span style='background:gainsboro'><font color='dimgray'>passengers</font></span>, donde si la observación es menor a la media de <span style='background:gainsboro'><font color='dimgray'>passengers</font></span> se le asigne a <span style='background:gainsboro'><font color='dimgray'>underperforming</font></span> un 1.

- Asignar unos en la columna <span style='background:gainsboro'><font color='dimgray'>underperforming</font></span> ya creada, para ello utilice la función <span style='background:gainsboro'><font color='dimgray'>at</font></span> de pandas.

**(3 puntos)**


```python
import pandas as pd
df= pd.read_csv('flights.csv')
df.head()
```

```python
#Objeto que representa a la media de 'passengers'
passengers_media = df['passengers'].mean()
passengers_media
```

```python
#Genera una columna en la base de datos que se llame underperforming y asígnele 0.
df["underperforming"]=0
df.head()
```

```python
#Si la observación es menor a la media de passengers se le asigne a underperforming un 1.
#Asignar unos en la columna underperforming ya creada, para ello utilice la función at de pandas

for indice, contenido_fila in df.iterrows():
    if contenido_fila ['passengers'] < passengers_media:
        df.at[indice, "underperforming"] =1
        
df
```

<font color='MediumSeaGreen'>**6.**</font> Por último, haz un loop que clasifique los meses donde la cantidad de pasajeros se escapa de la tendencia. Para ello:


- Genera dos objetos que guarden la media general y la desviación estándar general de <span style='background:gainsboro'><font color='dimgray'>passengers</font></span>.

- Genera una nueva columna en la tabla de datos que se llame <span style='background:gainsboro'><font color='dimgray'>outlier</font></span> y asígnale 0.

- Ejecuta un loop que recorra cada observación de <span style='background:gainsboro'><font color='dimgray'>passengers</font></span>, donde si la observación se escapa de la tendencia sea 1, de lo contrario 0.

- Para clasificar los casos que se escapen la tendencia, la observación debe satisfacer una de las siguientes condiciones:

    - La observación debe ser menor a la media menos la desviación estándar, o

    - La observación debe ser mayor a la media más la desviación estándar.

    - ¿Cuántas observaciones se pueden clasificar como casos extremos?

**(3 puntos)**


```python
#Dos objetos que guarden la media general y la desviación estándar general de passengers

passengers_media=df['passengers'].mean()
passengers_std=df['passengers'].std()

print(f"""Pasajeros:
Media: {passengers_media}
Desviación estándar: {passengers_std}""")
```

```python
#Genera una nueva columna en la tabla de datos que se llame outlier y asígnale 0.

df["outlier"]=0
df.head()
```

```python
#Si la observación se escapa de la tendencia sea 1, de lo contrario 0.
##caso_1 = La observación debe ser menor a la media menos la desviación estándar, o
##caso_2 = La observación debe ser mayor a la media más la desviación estándar.

caso_1=passengers_media-passengers_std
caso_2=passengers_media+passengers_std

print(f'caso 1= {caso_1}\ncaso 2= {caso_2}')

for indice, contenido_fila in df.iterrows():
    if contenido_fila ['passengers'] <caso_1 or contenido_fila ['passengers'] >caso_2:
        df.at[indice, "outlier"] =1
    else:
        df.at[indice, "outlier"] =0

df.head(10)        
```

```python
#contador de casos dentro de rango y casos extremos
df['outlier'].value_counts()
```

**¿Cuántas observaciones se pueden clasificar como casos extremos?**

- Se pueden clasificar *53 observaciones* como casos extremos en la desviación estandar, según los parametros dados.
